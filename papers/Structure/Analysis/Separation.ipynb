# Separation 

```python
from importlib import reload
import numpy as np
from cugn import arcteryx

import seaborn as sns
```

# Load

```python
adict = arcteryx.load_ctd()
```

# Cut on time

```python
reload(arcteryx)
dt, idx = arcteryx.cut_on_time(12., adict=adict)
```

```python
idx[0][0], idx[1][0]
```

# Calculate some separations

## Unpack for convenience

```python
d0 = adict['dist'][idx[0]]
d1 = adict['dist'][idx[1]]
#
o0 = adict['offset'][idx[0]]
o1 = adict['offset'][idx[1]]
```

```python
r = np.sqrt((d1-d0)**2 + (o0-o1)**2)
```

```python
ax = sns.histplot(r, bins=50, log_scale=True)
ax.set_xlabel('Separation (km)')
```
