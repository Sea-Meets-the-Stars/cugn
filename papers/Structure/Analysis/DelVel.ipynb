# Velocity

```python
from importlib import reload
import numpy as np
from cugn import arcteryx

import seaborn as sns
```

# Load

```python
adict = arcteryx.load_ctd()
```

# Cut on time

```python
reload(arcteryx)
dt, idx = arcteryx.cut_on_time(12., adict=adict)
```

# Separation (and r)

```python
reload(arcteryx)
rdict = arcteryx.calc_r(adict, idx)
```

# Velocity at a given depth

```python
iz = 0
# Unpack
u0 = adict['udop'][iz][idx[0]]
u1 = adict['udop'][iz][idx[1]]
v0 = adict['vdop'][iz][idx[0]]
v1 = adict['vdop'][iz][idx[1]]
```

```python
umag = np.sqrt((u1-u0)**2 + (v1-v0)**2)
```

## Examine

```python
sns.histplot(umag, bins=50)
```

# $\delta u_L$

```python
du = u1-u0
dv = v1-v0
```

```python
duL = rdict['dN']*du + rdict['oN']*dv
```

```python
sns.histplot(duL, bins=50)
```

```python
ax = sns.histplot(x=rdict['r'], y=np.abs(duL), log_scale=True)
#ax.set_xscale('log')
```

```python
gd = np.isfinite(duL)
```

```python
ax= sns.histplot(rdict['r'][gd], bins=50, log_scale=True)
#ax.set_xscale('log')
```
