# Compare All Ship and to AMSR2 SST (Microwave)

```python
import os
import glob
from importlib import reload

import numpy as np

import xarray
import pandas
import healpy

from matplotlib.ticker import FuncFormatter
from matplotlib import pyplot as plt
import seaborn as sns

from remote_sensing.healpix import rs_healpix
from remote_sensing.plotting import utils as plot_utils
from remote_sensing import io as rs_io

import ship
```

# Load

```python
jdict = rs_io.loadjson('Compare_SST_AMSR2_ship.json')
jdict.keys()
```

# Unfurl

```python
hours = []
medDTs = []
for key in jdict.keys():
    #
    hours.append(jdict[key]['hour'])
    medDTs.append(jdict[key]['medianDT'])
```

# Plot me

```python
fig = plt.figure(figsize=(11,7))
ax = plt.gca()

ax.plot(hours, medDTs, 'o')
#
ax.axhline(0., color='k', ls='--')
ax.set_xlabel('Hour (Local)')
ax.set_ylabel(r'$\Delta T$ (K) [Ship(inlet) - AMSR2]')
#ax.xaxis.set_major_formatter(FuncFormatter(timedelta_to_hours))

ax.set_ylim(-2., 2.)

plot_utils.set_fontsize(ax, 16.)
ax.grid()

plt.show()
```
